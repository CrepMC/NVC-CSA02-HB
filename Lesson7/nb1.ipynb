```python
import numpy as np
import matplotlib.pyplot as plt

# Dữ liệu mẫu
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 7, 11])

# Tính toán hệ số a và b
x_mean = np.mean(x)
y_mean = np.mean(y)

# Công thức tính a
numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)
a = numerator / denominator

# Công thức tính b
b = y_mean - a * x_mean

print(f"Hàm số hồi quy: y = {a:.2f}x + {b:.2f}")

# Dự đoán
y_pred = a * x + b

# Vẽ đồ thị
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color="red", s=100, label="Dữ liệu gốc")
plt.plot(x, y_pred, color="blue", linewidth=2, label=f"Hồi quy: y = {a:.2f}x + {b:.2f}")
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.title("Hồi quy tuyến tính đơn", fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Tính MSE
mse = np.mean((y - y_pred) ** 2)
print(f"Mean Squared Error: {mse:.4f}")
# Tính R-squared
y_mean = np.mean(y)
ss_tot = np.sum((y - y_mean) ** 2)   
ss_res = np.sum((y - y_pred) ** 2)
r_squared = 1 - (ss_res / ss_tot)
print(f"R-squared: {r_squared:.4f}")
```
