```python
import pandas as pd
```

```python
df = pd.read_csv('Billionaires.csv')
```

### The function is to remove B and $ of net worth object

```python
def clean_networth(nw):
    if isinstance(nw, str):
        return nw[1:-1]
    return nw

df['Net Worth'] = df['Net Worth'].apply(clean_networth)
df['Net Worth'] = pd.to_numeric(df['Net Worth'])
df.head()
```

```python
df[["Industry", "Net Worth"]].groupby(["Industry"]).sum().sort_values(by="Net Worth", ascending=False)
```

```python
import pycountry_convert as pc

def get_continient(country_name):
    try:
        country_code = pc.country_name_to_country_alpha2(country_name)
        continient_code = pc.country_alpha2_to_continent_code(country_code)
        continient_name = pc.convert_continent_code_to_name(continient_code)
        return continient_name
    except:
        print("Error converting country name:", country_name)
        return "Unknown"

df["Continent"] = df["Country | Territory"].apply(get_continient)
```

```python
df["Continient"] = df["Country | Territory"].apply(get_continient)
```

```python
df[["Continient", "Net Worth"]].groupby(["Continent"]).sum().sort_values(by="Net Worth", ascending=False)
```
