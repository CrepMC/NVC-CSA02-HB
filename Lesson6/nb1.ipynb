```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
spotify_data = pd.read_csv('spotify.csv')
spotify_data['Date'] = pd.to_datetime(spotify_data['Date'], format='%Y-%m-%d')
```

```python
plt.figure(figsize=(14, 6))

df_line = spotify_data.set_index('Date')
plt.plot(df_line)

plt.title('Spotify Data Over Time')
plt.xlabel('Date')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.legend(df_line.columns, loc='upper left')
plt.tight_layout()
plt.savefig('spotify_data_plot.png')
```
