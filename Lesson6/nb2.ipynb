```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Billionaires.csv')

def clean_networth(nw):
        return nw[1:-2]

df['Net Worth'] = df['Net Worth'].apply(clean_networth)
df['Net Worth'] = pd.to_numeric(df['Net Worth'])
df.head()
```

```python
df_top10 = df[["Industry", "Net Worth"]].groupby("Industry").sum().sort_values("Net Worth", ascending=False)[0:10]

plt.figure(figsize=(14, 6))

plt.bar(x=df_top10.index, height=df_top10["Net Worth"])
plt.xticks(rotation=90)
plt.title("Top 10 Industries by Total Net Worth of Billionaires")
plt.xlabel("Industry") 
plt.ylabel("Net Worth")
```

```python
plt.figure(figsize=(8, 4))
plt.hist(x=df["Age"], edgecolor='white')
plt.title("Age Distribution of Billionaires")
plt.xlabel("Age")
plt.ylabel("Number of Billionaires")
```

```python
df_top10 = df[["Age", "Net Worth"]]

plt.figure(figsize=(6, 6))

plt.scatter(x=df_top10["Age"], y=df_top10["Net Worth"])
plt.xticks(rotation=90)
plt.title("Top 10 Industries by Total Net Worth of Billionaires")
plt.xlabel("Age") 
plt.ylabel("Net Worth")
```
